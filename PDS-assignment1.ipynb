Exercise: Replace the missing values in numeric columns of  dataset (hepatitis dataset) with the mean values of the column.

Importing the libraries

```
import pandas as pd
import numpy as np
```

reading the hepatitis dataset from Github to do the assignment 1 tasks on it



```
df = pd.read_csv('https://raw.githubusercontent.com/datasets/hepatitis/master/data/hepatitis.csv')
df.head(10)
```

df.select_dtypes helps to select only the columns in the DataFrame (df) that have numeric data types (float64 or int64) and .columns returns the names of these selected columns as an index object (essentially, a list of column names that are numeric). for col in numeric_cols: The for loop loops through each numeric column in numeric_cols. And then mean_value calculates the mean (average) of the current column (col). If the column has missing values (NaN), they are ignored when calculating the mean. The .fillna replaces all the missing values (NaN) in the current column with the mean value of that column and results a column with no missing values.



```
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
for col in numeric_cols:
    mean_value = df[col].mean()
    df[col] = df[col].fillna(mean_value)
df.head()
```
